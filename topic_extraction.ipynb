```python
import string
import nltk
import numpy as np

from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.util import ngrams

# nltk.download()
```

```python
fileName = './TFaS_p108.txt'
testFile = open(fileName, 'r')
testPage = testFile.read()
testFile.close()
```

```python
# new paragraphs start with a newline and 4 spaces
raw_paragraphs = testPage.split("\n    ")
```

```python
# make it numpy
#raw_paragraphs = np.asarray(raw_paragraphs)
```

```python
# clean newline characters
raw_paragraphs = [paragraph.replace('\n', ' ') for paragraph in raw_paragraphs] #np.array()

# tokenize
# TODO: consider the impact of punctuation on the output (the following lines remove it)
# TODO: couldn't get r'\w+|\d+[,.]?(\d+[,.]?)+' to work, now decimal numbers 
tokenizer = RegexpTokenizer(r'\w+')
raw_paragraphs = [tokenizer.tokenize(paragraph) for paragraph in raw_paragraphs] #np.array()
```

```python
# remove stopwords
# TODO: stopwords like 'is' sometimes provide useful unformation. It can be skipped for now for topic modelling though
stopWords = set(stopwords.words('english'))
raw_paragraphs = [[word for word in paragraph if word not in stopWords] for paragraph in raw_paragraphs] #np.array()
```

```python
# lemmatize
lemmatizer = nltk.stem.WordNetLemmatizer()
raw_paragraphs = [[lemmatizer.lemmatize(w) for w in paragraph] for paragraph in raw_paragraphs] # np.asarray(  np.asarray()
```

```python
raw_paragraphs
```

```python
prepared_paragraphs = []

for paragraph in raw_paragraphs:
    n = 2
    ngram_generator = ngrams(paragraph, n)
    prepared_paragraphs.append([' '.join(ngrams) for ngrams in ngram_generator ])
```
